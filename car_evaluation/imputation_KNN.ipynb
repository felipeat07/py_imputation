```python
import pandas as pd
import numpy as np
```

```python
# Caminho para o arquivo CSV
file_path = 'car_data/car.data'

# Lista correta com os nomes das colunas, conforme a descrição do dataset Car Evaluation
columns = [
    'buying', 'maint', 'doors', 'persons', 'lug_boot', 'safety', 'class'
]

# Leitura do arquivo
df = pd.read_csv(file_path, header=None, names=columns)

# Configurar o pandas para mostrar todas as colunas
pd.set_option('display.max_columns', None)

# Exibir as primeiras 10 linhas com todas as colunas
display(df.head(100))
```

```python
# Verifica dados ausentes (valores NaN)
print(df.isnull().sum())

# Também verifica se há valores representando ausência como '?'
print(df.eq('?').sum())
```

```python
# Contar número de categorias únicas por coluna
categorias_unicas = df.nunique()
display(categorias_unicas)
```
