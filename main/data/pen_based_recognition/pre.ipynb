```python
import pandas as pd

# Colunas do dataset
cols = [
    'X1', 'Y1', 'X2', 'Y2', 'X3', 'Y3',
    'X4', 'Y4', 'X5', 'Y5', 'X6', 'Y6',
    'X7', 'Y7', 'X8', 'Y8', 'Class'
]

# Leitura dos dois arquivos
train = pd.read_csv('pendigits.tra', header=None, names=cols)
test  = pd.read_csv('pendigits.tes', header=None, names=cols)

# Junta os dois
df = pd.concat([train, test], ignore_index=True)

# Salva como CSV
df.to_csv('pendigits_full.csv', index=False)

print("Arquivo salvo: pendigits_full.csv")
print("Linhas:", len(df))
print("Colunas:", df.columns.tolist())
```
