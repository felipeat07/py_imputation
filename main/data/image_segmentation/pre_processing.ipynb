```python
import pandas as pd

COLUMNS = [
    'Class',
    'Region-centroid-col', 'Region-centroid-row',
    'Region-pixel-count', 'Short-line-density-5',
    'Short-line-density-2', 'Vedge-mean',
    'Vedge-sd', 'Hedge-mean', 'Hedge-sd',
    'Intensity-mean', 'Rawred-mean',
    'Rawblue-mean', 'Rawgreen-mean',
    'Exred-mean', 'Exblue-mean', 'Exgreen-mean',
    'Value-mean', 'Saturation-mean', 'Hue-mean'
]

# Carregar treino
df_train = pd.read_csv(
    'segmentation.data',
    sep=',',
    header=None,
    names=COLUMNS,
    skiprows=3  # ignora cabeçalho extra
)

# Carregar test
df_test = pd.read_csv(
    'segmentation.test',
    sep=',',
    header=None,
    names=COLUMNS,
    skiprows=3
)

# Combinar
df = pd.concat([df_train, df_test], ignore_index=True)

# Normalizar target
df['Class'] = df['Class'].astype(str).str.strip().str.lower()

# Verificar duplicatas / sujeira
print("Classes antes de limpar:")
print(df['Class'].value_counts())

# O dataset não tem valores ausentes por definição :contentReference[oaicite:2]{index=2}

# Salvar o CSV completo e limpo
df.to_csv('egmentation_full_clean.csv', index=False)
print("Salvo com sucesso!")
print("Shape final:", df.shape)
print(df['Class'].value_counts())
```
