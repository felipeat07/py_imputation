```python
import pandas as pd

# Caminho do arquivo CSV
file_path = 'Students_data/Students Social Media Addiction.csv'

# Carregar o dataset com cabeçalho do próprio arquivo
df = pd.read_csv(file_path)

# Exibir as 10 primeiras linhas
display(df.head(10))

# Verificar valores nulos por coluna
missing_values = df.isnull().sum()
print("Valores nulos por coluna:\n", missing_values)

# Total de valores nulos
print("\nTotal de valores nulos:", missing_values.sum())


# Verificar valores ausentes representados por interrogação '?'
missing_question = (df == '?').sum()
print("\nValores representados por '?':\n", missing_question)
print("\nTotal de '?':", missing_question.sum())
```

```python
# Lista das colunas categóricas
categorical_columns = [
    'Gender', 
    'Academic_Level', 
    'Country', 
    'Most_Used_Platform', 
    'Affects_Academic_Performance', 
    'Relationship_Status'
]

print("Número de categorias distintas por coluna:")
for col in categorical_columns:
    num_unique = df[col].nunique(dropna=True)  # dropna=True ignora NaN
    print(f"{col}: {num_unique} categoria(s)")
```

```python
# Lista das colunas categóricas
categorical_columns = [
    'Gender', 
    'Academic_Level', 
    'Country', 
    'Most_Used_Platform', 
    'Affects_Academic_Performance', 
    'Relationship_Status'
]

# Verificar distribuição percentual por coluna
for col in categorical_columns:
    print(f"\n--- {col} ---")
    print(df[col].value_counts(normalize=True, dropna=False).apply(lambda x: f"{x:.2%}"))
```
